```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

codes = "ex_cpu_double/long_sim/note","texture_double_precision/long_sim/note"

x,sum_prog,err_prog = np.loadtxt(codes[0],usecols=(1,2,3),unpack='true')
plt.errorbar(x,sum_prog,yerr=err_prog)
x,sum_prog,err_prog = np.loadtxt(codes[1],usecols=(1,2,3),unpack='true')
plt.errorbar(x,sum_prog,yerr=err_prog,marker='o',linestyle=' ')
plt.xlabel("density",fontsize=15)
plt.ylabel("pressure",fontsize=15)
plt.legend(codes)


plt.savefig("trans_phase.png") 
plt.show()

codes = "ex_cpu_double/long_sim/","texture_double_precision/long_sim/"

phase = '1','07','05','025','02','01'
name = [" rho 1"," rho 07"," rho 05"," rho 025"," rho 02"," rho 01"]
fig,axs = plt.subplots(nrows=2,ncols=3,figsize=(25,20))

k=0
for i in range(2):
    for j in range(3):
        for l in range(2):
            y = np.loadtxt(codes[l]+"output.gave"+phase[k]+".out",usecols=(0),unpack='true')
            sum_prog,err_prog = np.loadtxt(codes[l]+"output.gave"+phase[k]+".out",usecols=(1,2),unpack='true')
            axs[i][j].errorbar(y,sum_prog,yerr=err_prog)
        axs[i][j].set_xlabel("r*",fontsize=20)
        axs[i][j].set_ylabel("$g(r)$",fontsize=20)
        axs[i][j].set_title("pair radial correlation function"+name[k],fontsize=20)
        axs[i][j].legend(codes)
        k+=1
            

            
plt.savefig("trans_phase_gr.png")

plt.show()
```
